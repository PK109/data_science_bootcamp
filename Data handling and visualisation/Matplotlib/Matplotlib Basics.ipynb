```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
# functional creating of plots
plt.plot(x,y, 'r-')
plt.xlabel("X Label")
plt.ylabel('Y Label')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y)
plt.subplot(1,2,2)
plt.plot(y,x,'r')
```

```python
#Object oriented creating of plots
fig = plt.figure()

axes = fig.add_axes([0.2,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_ylabel('Y label')
axes.set_xlabel('X label')
axes.set_title('Title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.5,0.6,0.4,0.3])
axes2.set_ylabel('Y label')
axes2.set_xlabel('X label')
axes2.set_title('SMALLER')
```

```python
fig, axes = plt.subplots(nrows=3,ncols=3)
#axes.plot(x,y)
plt.tight_layout()
for row in axes:
    for graph in row:
        graph.plot(x,y)
```

# Fig Size and DPI

```python
fig = plt.figure(figsize=(6,3))
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.plot(y,x)
ax.set_ylabel('Y label')
ax.set_xlabel('X label')
ax.set_title('Title')
ax.legend(['X square','X sqrt'],loc= 0)
```

```python
fig, axes = plt.subplots(figsize=(6,3), nrows = 2)
axes[0].plot(x,y)
axes[1].plot(y,x)
```

# Plot Appearance

```python
fig = plt.figure(figsize=(6,3))
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color = 'orange', linewidth=3, linestyle=':') #lw lub ls robi to samo
ax.plot(y,x,color = '#FF00C0', alpha= 0.3, marker = 'o', markersize = 20, markerfacecolor = 'yellow') #rgb hex code
```

```python
fig, ax = plt.subplots(figsize=(12,6))

ax.plot(x, x+1, color="red", linewidth=0.25)
ax.plot(x, x+2, color="red", linewidth=0.50)
ax.plot(x, x+3, color="red", linewidth=1.00)
ax.plot(x, x+4, color="red", linewidth=2.00)

# possible linestype options -, , -., :, steps
ax.plot(x, x+5, color="green", lw=3, linestyle='-')
ax.plot(x, x+6, color="green", lw=3, ls='-.')
ax.plot(x, x+7, color="green", lw=3, ls=':')

# custom dash
line, = ax.plot(x, x+8, color="black", lw=1.50)
line.set_dashes([5, 10, 15, 10]) # format: line length, space length, ...

# possible marker symbols: marker = '+', 'o', '*', 's', ',', '.', '1', '2', '3', '4', ...
ax.plot(x, x+ 9, color="blue", lw=3, ls='-', marker='+')
ax.plot(x, x+10, color="blue", lw=3, ls='--', marker='o')
ax.plot(x, x+11, color="blue", lw=3, ls='-', marker='s')
ax.plot(x, x+12, color="blue", lw=3, ls='--', marker='1')

# marker size and color
ax.plot(x, x+13, color="purple", lw=1, ls='-', marker='o', markersize=2)
ax.plot(x, x+14, color="purple", lw=1, ls='-', marker='o', markersize=4)
ax.plot(x, x+15, color="purple", lw=1, ls='-', marker='o', markersize=8, markerfacecolor="red")
ax.plot(x, x+16, color="purple", lw=1, ls='-', marker='s', markersize=8, 
        markerfacecolor="yellow", markeredgewidth=3, markeredgecolor="green");
```
