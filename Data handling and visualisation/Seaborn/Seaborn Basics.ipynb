```python
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

# Distribution Plots

```python
sns.displot(tips['total_bill'],kde= True, bins = 20)
```

```python
sns.jointplot(x='total_bill' , y='tip' , data= tips, kind='reg')
```

```python
sns.pairplot(tips, hue= 'sex')
```

```python
sns.rugplot(tips['total_bill'])
```

# Categorical Plots

```python
import numpy as np
```

```python
np.std
```

```python
sns.barplot(x='sex', y='total_bill', data=tips, estimator= np.std)
```

```python
sns.countplot(x='sex', data=tips)
```

```python
sns.boxplot(x='day',y='total_bill', data=tips, hue='sex')
```

```python
sns.violinplot(x='day',y='total_bill', data=tips, hue='sex', split=True)
```

```python
sns.stripplot(x='day',y='total_bill', data=tips, hue='sex', dodge=True)
```

```python
sns.swarmplot(x='day',y='total_bill', data=tips, hue='sex', dodge=True, size=3)
```

```python
sns.catplot(x='day',y='total_bill', data=tips, kind='boxen')
```

# Grids

```python
import matplotlib.pyplot as plt
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
sns.pairplot(iris)
```

```python
g = sns.PairGrid(iris)
g.map_diag(sns.histplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
g = sns.FacetGrid(data=tips, col='time', row='smoker')
g.map(sns.histplot, 'total_bill', kde=True)
```

# Regression Plot

```python
sns.lmplot(x='total_bill',y='tip',data=tips, hue='sex', markers=['o','v'], scatter_kws={'s':50})
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips, col='day',hue='sex', aspect= 0.6, height=8)
```
